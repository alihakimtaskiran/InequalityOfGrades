**Table 1:**Letter grade assignment table



```python
def grader(arg):
    if arg == 'A+':return 4.0
    elif arg == 'A':return 4.0
    elif arg == 'A-':return 3.7
    elif arg == 'B+':return 3.3
    elif arg == 'B':return 3.0
    elif arg == 'B-':return 2.7
    elif arg == 'C+':return 2.3
    elif arg == 'C':return 2.0
    elif arg == 'C-':return 1.7
    elif arg == 'D+':return 1.3
    elif arg == 'D':return 1.0
    elif arg == 'F' or arg == 'FX' or arg == 'FZ':return 0.0
    else:return None
```

```python
def gini(arg):
    cum_people=[0]
    cum_wealth=[0]

    for i in arg:
        if i!='name':
            cum_people.append(cum_people[-1]+arg[i])
            cum_wealth.append(cum_wealth[-1]+arg[i]*grader(i))

    area_of_equality=np.trapz((0,cum_people[-1]),(0,cum_wealth[-1]))
    b=sum(cum_wealth)
    cum_people= np.array(cum_people)
    cum_wealth=np.array(cum_wealth)
    b=np.trapz(cum_people,cum_wealth)
    return cum_people, cum_wealth, (b-area_of_equality)/area_of_equality
```

## Gini Index of Courses
The Gini index is a statistical measure commonly used to quantify the degree of income inequality or wealth distribution within a population. The index ranges between 0 and 1, where 0 represents perfect equality (every individual has an equal share of income or wealth), and 1 represents maximum inequality (one individual has all the income or wealth, while others have none).

Since reference intervals are typically used for laboratory test results, it is important to note that the Gini index is not a laboratory test, but rather a statistical measure. As such, it does not have standard reference intervals like clinical or laboratory parameters.

Instead, the interpretation of the Gini index relies on the context and the specific population under consideration. In general, the following guidelines can be used as a reference:

- A Gini index below 0.2 indicates a relatively low level of inequality in course-related characteristics. This suggests that there is minimal disparity among individuals in terms of course participation, achievement, or other measured factors. The distribution of these characteristics is more equal, with a smaller gap between different individuals.

-   A Gini index ranging from 0.2 to 0.3 suggests a moderate level of inequality. While some variations may exist in course-related characteristics, the distribution is still relatively equitable. The disparities among individuals are not significant, indicating a relatively lower level of inequality compared to higher Gini index ranges.

-   Gini index values between 0.3 and 0.4 represent a relatively reasonable income gap or disparity in course-related characteristics. There are noticeable differences among individuals in terms of course enrollment, performance, or other relevant factors, but the level of inequality is not highly pronounced.

-   Gini index values between 0.4 and 0.5 correspond to a higher level of income disparity or inequality in course-related characteristics. In this range, the variations among individuals become more pronounced, and a larger share of the outcomes is concentrated among a smaller portion of the population.

-   Gini index values above 0.5 indicate a severe level of income disparity or inequality in course-related characteristics. The distribution becomes highly skewed, with a smaller fraction of individuals dominating the course enrollment, performance, or other measured factors, while the majority experiences limited access or success.

It is important to note that the interpretation of the Gini index should be done with caution, as it does not provide a comprehensive picture of inequality. Other socioeconomic factors, such as access to education, healthcare, and opportunities, should be considered alongside the Gini index to have a more complete understanding of the distribution of income or wealth within a population.

## Lorenz Curve Plots

```python
normalized_database={}
for i in database:
    x,y,g=gini(i)
    normalized_database[i['name']]=x/x[-1]*100,y/y[-1]*100,g
colors = [
'red', 'blue', 'green', 'yellow', 'orange', 'purple', 'pink', 'brown', 'gray', 'black', 'deepskyblue', 'cyan', 'magenta', 'teal', 'maroon', 'navy', 'olive', 'lime', 'aqua', 'silver', 'gold', 'violet', 'indigo', 'coral', 'turquoise', 'salmon'
]

plt.figure(figsize=(20,10))
plt.title("Lorenz Curves of Various Courses in Bilkent University")
cc=0
plt.plot((0,100),[0,4],label='Line of Equality',color='maroon')

for i in normalized_database:
    x,y,g=normalized_database[i]
    plt.plot(x,y/25,'.--',label=i,color=colors[cc])
    cc+=1

plt.xlabel('Number of Agents(%)')
plt.ylabel('Cumulative Average Grade')
plt.legend()
plt.show()
```
